import library

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
import pickle
```

load datasetnya

```python
dataset = pd.read_csv('Dataset_A_loan.csv')
```

```python
dataset.head(10)
```

```python
dataset.tail(10)
```

```python
dataset.isnull().sum()
```

mengonversi kolom kategorikal jadi numerik

```python
label_encoder = LabelEncoder()
dataset['person_gender'] = label_encoder.fit_transform(dataset['person_gender'])
dataset['person_education'] = label_encoder.fit_transform(dataset['person_education'])
dataset['person_home_ownership'] = label_encoder.fit_transform(dataset['person_home_ownership'])
dataset['loan_intent'] = label_encoder.fit_transform(dataset['loan_intent'])
dataset['previous_loan_defaults_on_file'] = label_encoder.fit_transform(dataset['previous_loan_defaults_on_file'])
```

memisahkan data fitur dan target

```python
X = dataset.drop('loan_status', axis=1)
y = dataset['loan_status']
```

menangani missing values

```python
imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X)
```

membagi dataset jadi data training dan test

```python
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)
```

melatih model pertama, random forest

```python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

melatih model kedua, xgboost

```python
xgb_model = XGBClassifier(random_state=42)
xgb_model.fit(X_train, y_train)
```

membuat prediksinya

```python
rf_pred = rf_model.predict(X_test)
xgb_pred = xgb_model.predict(X_test)
```

mengevaluasi kedua model

```python
rf_accuracy = accuracy_score(y_test, rf_pred)
xgb_accuracy = accuracy_score(y_test, xgb_pred)

rf_classification_report = classification_report(y_test, rf_pred)
xgb_classification_report = classification_report(y_test, xgb_pred)
```

menampilkan hasilnya

```python
print(f"Akurasi dari Random forest : {rf_accuracy}")
print(f"Akurasi dari XGBoost: {xgb_accuracy}")

print("Classification Report dari Random forest:")
print(rf_classification_report)
print("Classification Report dari XGBoost:")
print(xgb_classification_report)
```

menyimpan model terbaik yaitu xgboost (dipilih karena hasil evaluasinya)

```python
best_model = xgb_model
pickle.dump(best_model, open('best_model.pkl', 'wb'))
```
